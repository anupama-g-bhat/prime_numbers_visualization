```python
import matplotlib.pyplot as plt 

def is_prime(n):
    primes = []
    j = 2
    while len(primes) < n:
        for i in range(2,j):
            if(j % i == 0):
                break
        else :
            primes.append(j)    
        j += 1         

    return primes

#Taking input
n = int(input("Enter a number to find first n prime numbers :"))

#Getting primes
primes = is_prime(n)

#Creating x-axis value
x = list(range(n))

# Create color list based on value
colors = [ "blue" if num<50 else "red" if num<100  else "green" for num in primes]

#Visualization
plt.scatter(x,primes,c = colors, s=40)
plt.plot(x,primes,linestyle="--",color="gray")
plt.title("Scatter Plot of First N Prime Numbers")
plt.xlabel("Index")
plt.ylabel("Prime Numbers")
plt.show()





```
